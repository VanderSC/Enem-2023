```python
#https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/sinopses-estatisticas/enem

import matplotlib.pyplot as plt
import numpy as np


# Dados
regioes = ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
total_inscritos = [454772, 1424505, 1305607, 416416, 332655]

total_inscritos_formatados = [f"{total:,}" for total in total_inscritos]

plt.figure(figsize=(10, 6))
bars = plt.bar(regioes, total_inscritos, color='red')

plt.title('Número Total de Inscritos no ENEM por Região Geográfica')
plt.xlabel('Região Geográfica')
plt.ylabel('Número de Inscritos')

for bar, total in zip(bars, total_inscritos_formatados):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 10000, total, fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Dados
regioes = ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
masculino = [175517, 557023, 500711, 158594, 130925]
feminino = [279255, 867482, 804896, 257822, 201730]

bar_width = 0.35
index = np.arange(len(regioes))

plt.figure(figsize=(10, 6))
plt.bar(index, masculino, bar_width, label='Masculino', color='blue')
plt.bar(index + bar_width, feminino, bar_width, label='Feminino', color='red')

plt.xlabel('Região Geográfica')
plt.ylabel('Número de Inscritos')
plt.title('Distribuição de Inscritos no ENEM por Sexo e Região Geográfica')
plt.xticks(index + bar_width / 2, regioes)
plt.legend()

plt.tight_layout()
plt.show()
```
